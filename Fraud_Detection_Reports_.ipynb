# Banking Transaction Analysis and Fraud Detection Reports

### 1. Transaction Volume Over Time (step)

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')

# Assuming 'step' is already in a format that can be used to represent time units directly
transactions_over_time = data.groupby('step')['amount'].count()

# Plotting
transactions_over_time.plot(kind='line')
plt.title('Transaction Volume Over Time')
plt.xlabel('Step')
plt.ylabel('Number of Transactions')
plt.show()
```

###  2. Total Amount Transacted Over Time (step)

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')

amount_over_time = data.groupby('step')['amount'].sum()

# Plotting
amount_over_time.plot(kind='line')
plt.title('Total Amount Transacted Over Time')
plt.xlabel('Step')
plt.ylabel('Total Amount')
plt.show()
```

### 3. Distribution of Transaction Types (type)

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')

transaction_types = data['type'].value_counts()

# Plotting
transaction_types.plot(kind='bar')
plt.title('Distribution of Transaction Types')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.show()
```

#### 4. Fraudulent Transactions by Type (type, isFraud)

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')

fraud_by_type = data[data['isFraud'] == 1]['type'].value_counts()

# Plotting
fraud_by_type.plot(kind='bar')
plt.title('Fraudulent Transactions by Type')
plt.xlabel('Type')
plt.ylabel('Number of Fraudulent Transactions')
plt.show()
```

#### 5. Flagged vs Actual Fraud Incidents (isFraud, isFlaggedFraud)

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')

fraud_comparison = data[['isFraud', 'isFlaggedFraud']].sum()

# Plotting
fraud_comparison.plot(kind='bar')
plt.title('Flagged vs Actual Fraud Incidents')
plt.xticks(ticks=[0, 1], labels=['Actual Fraud', 'Flagged Fraud'], rotation=0)
plt.ylabel('Count')
plt.show()
```

#### 6. Changes in Balance for Origin Accounts (oldbalanceOrg, newbalanceOrig)

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')

data['balance_change_orig'] = data['newbalanceOrig'] - data['oldbalanceOrg']

# Analyze large changes
large_changes_orig = data[data['balance_change_orig'].abs() > data['balance_change_orig'].quantile(0.99)]

# Summary
large_changes_summary = large_changes_orig['balance_change_orig'].describe()
print(large_changes_summary)
```
